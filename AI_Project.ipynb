```python
import pandas as pd  # For working with structured data using DataFrames and Series
from sklearn.model_selection import train_test_split   # For splitting the dataset into training and testing sets
from sklearn.linear_model import LogisticRegression    # Import the Logistic Regression classifier from scikit-learn
from sklearn.metrics import accuracy_score             # For calculating the accuracy of the model's predictions
```

```python
from google.colab import drive  # Used to mount Google Drive in Google Colab to access datasets stored on Drive
drive.mount("/content/drive")
```

```python
load = pd.read_csv("/content/drive/MyDrive/heart .csv") # variable to store dataset
load  # print dataset
```

```python
load.info() # print number of rows , columns , name of columns , datatype of column , missing values and memory usage
```

```python
X = load.drop("Result", axis=1) # delete column of target and store columns of features
y = load["Result"] # store column of target
y # print column of target
```

```python
y.value_counts() # Shows the number of samples in each class of the target variable
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# Splitting the data into:
# X_train, y_train → Training data (70%)
# X_test, y_test → Test data (30%)
# test_size=0.3 means 30% of the data is for testing
# random_state=42 ensures the split produces the same result every time we run the code
```

```python
model = LogisticRegression(solver='liblinear', # The appropriate algorithm for training binary classification
                           max_iter=700, # Maximum number of training attempts
                           verbose=1 # Show details during training
                           )
model.fit(X_train, y_train) # The logistic regression model was trained using the training dataset

y_pred = model.predict(X_test) # The trained model is used to predict the target values for the test data

accuracy = accuracy_score(y_test, y_pred)  # Accuracy measures the percentage of correctly predicted samples by the model
print("Accuracy:", accuracy) # To print accuracy
```
